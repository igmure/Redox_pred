```python
folder_path = "./sdf_2"
df12=pd.DataFrame()
dataset = DatasetToCSVROP313Class_sdf()
df = dataset.process_files(folder_path)
df12=df.dropna()
df12['system_number'].to_list()
print(len(df12['system_number'].to_list()))
```

```python

#calculate all desctiptors based on smiles
import pandas as pd 
from rdkit import Chem
from mordred import Calculator, descriptors
import numpy as np 

np.float = float

pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)
mols = df12['sdf']

calc=Calculator(descriptors, ignore_3D=False)

res = [calc(mol).asdict() for mol in mols]

df2=pd.DataFrame(res)

also_trash=[
    col for col in df2.columns
    if df2[col].dtypes =="object"
]

df2=df2.drop(columns=also_trash)

df2
#check if the descriptors correlate strongly with each other and remove such instances

corr_mat=df2.corr()

high_corr = np.where(np.abs(corr_mat) >= 0.8)
high_corr_pairs = [
    (corr_mat.index[x], corr_mat.columns[y])
    for x, y in zip(*high_corr) if x != y
]

to_rm=set()
for des1, des2 in high_corr_pairs:
    if des1 not in to_rm and des2 not in to_rm:
        avg_corr_1 = corr_mat[des1].mean()
        avg_corr_2 = corr_mat[des2].mean()
        if avg_corr_1 > avg_corr_2:
            to_rm.add(des1)
        else:
            to_rm.add(des2)
        
reduced_df = df2.drop(columns=to_rm)

trash=reduced_df.columns[(reduced_df == 0).all()]
reduced_df=reduced_df.drop(columns=trash)
also_trash=[
    col for col in reduced_df.columns
    if reduced_df[col].dtypes =="object"
]

clean_df=reduced_df.drop(columns=also_trash)

# clean_df
even_cleaner_df = pd.concat([pd.DataFrame(df12['system_number'].to_list(), columns=['system_number']), clean_df], axis=1)
even_cleaner_df

```

```python
even_cleaner_df.to_csv("sdf_descriptors.csv", index=False)
```
